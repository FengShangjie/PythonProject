```python
import os,re,csv,math

import pandas as pd
import numpy as np

zc = ['现金与约当现金(占总资产%)', '应收款项(占总资产%)', '存货(占总资产%)', '流动资产(占总资产%)', '非流动资产(占总资产%)',\
     '应付款项(占总资产%)', '流动负债(占总资产%)', '非流动负债(占总资产%)', '股东权益(占总资产%)',\
     '负债占资产比率(%)', '长期资金占不动产及设备比率(%)', '流动比率(%)', '速动比率(%)','应收款项周转率(次/年)', '应收款项周转天数(天)', \
      '存货周转率(次/年)', '存货周转天数(天)', '固定资产周转率(次/年)', '完整生意周期(天)','应付款项周转天数(天)', '缺钱天数(天)', \
      '总资产周转率(次/年)', 'ROA=资产收益率(%)', 'ROE=净资产收益率(%)', '税前纯益占实收资本比率(%)','毛利率(%)', '营业利润率(%)', '净利率(%)',\
     '营业费用率(%)','经营安全边际率(%)','EPS=基本每股收益(元)', '营收增长率(%)', '净利增长率(%)','净资本增长率(%)',\
     '现金流量比率(%)','现金流量允当比率(%)', '现金再投资比率(%)']

years=[2013,2014,2015,2016,2017]
gp =pd.DataFrame(columns=years, index=zc)
#----------------------------------------打开文件夹中对应的三张Excel表-----------------------------------    
path_root = '..\\OutFile\\A股报表大全'
dirs = os.listdir(path_root)
w = 0
for dir in dirs:
    w=w+1
    gp =pd.DataFrame(columns=years, index=zc)
    print("正在处理："+dir+'已处理'+str(w*100/len(dirs)))
    path_stock=os.path.join(path_root,dir)
    path_fzb=os.path.join(path_stock,'fzb')
    path_llb=os.path.join(path_stock,'llb')
    path_lrb=os.path.join(path_stock,'lrb')
    ls=dir.split('_')
    file_count = 0
    for fn in os.listdir(path_fzb): #部分公司上市少于5年，目录下的文件数少于5个，获得具体数量，为以下循环提供数据
        file_count = file_count + 1
    if file_count<len(years):
        years_cnt = file_count
    else:
        years_cnt=len(years)
    for i in range(years_cnt):
        l1 = []
        if ls[0].startswith('6'): #沪市A股代码以数字6打头
            path_fzb_year=os.path.join(path_fzb,'sh_fzb_'+ls[0]+'_'+str(years[i])+'.csv')
            path_fzb_pre_year = os.path.join(path_fzb,'sh_fzb_'+ls[0]+'_'+str(years[i]-1)+'.csv')
            path_llb_year=os.path.join(path_llb,'sh_llb_'+ls[0]+'_'+str(years[i])+'.csv')
            path_lrb_year=os.path.join(path_lrb,'sh_lrb_'+ls[0]+'_'+str(years[i])+'.csv')
            path_lrb_pre_year=os.path.join(path_lrb,'sh_lrb_'+ls[0]+'_'+str(years[i]-1)+'.csv')
            path_fzb_five_year = os.path.join(path_fzb,'sh_fzb_'+ls[0]+'_'+str(years[i]-4)+'.csv')
            path_llb_two_year=os.path.join(path_llb,'sh_llb_'+ls[0]+'_'+str(years[i]-1)+'.csv')
            path_llb_three_year=os.path.join(path_llb,'sh_llb_'+ls[0]+'_'+str(years[i]-2)+'.csv')
            path_llb_four_year=os.path.join(path_llb,'sh_llb_'+ls[0]+'_'+str(years[i]-3)+'.csv')
            path_llb_five_year=os.path.join(path_llb,'sh_llb_'+ls[0]+'_'+str(years[i]-4)+'.csv')
            
        else:
            path_fzb_year=os.path.join(path_fzb,'sz_fzb_'+ls[0]+'_'+str(years[i])+'.csv')
            path_fzb_pre_year = os.path.join(path_fzb,'sz_fzb_'+ls[0]+'_'+str(years[i]-1)+'.csv')
            path_llb_year=os.path.join(path_llb,'sz_llb_'+ls[0]+'_'+str(years[i])+'.csv')
            path_lrb_year=os.path.join(path_lrb,'sz_lrb_'+ls[0]+'_'+str(years[i])+'.csv')   
            path_lrb_pre_year=os.path.join(path_lrb,'sz_lrb_'+ls[0]+'_'+str(years[i]-1)+'.csv')   
            path_fzb_five_year = os.path.join(path_fzb,'sz_fzb_'+ls[0]+'_'+str(years[i]-4)+'.csv')
            path_llb_two_year=os.path.join(path_llb,'sz_llb_'+ls[0]+'_'+str(years[i]-1)+'.csv')
            path_llb_three_year=os.path.join(path_llb,'sz_llb_'+ls[0]+'_'+str(years[i]-2)+'.csv')
            path_llb_four_year=os.path.join(path_llb,'sz_llb_'+ls[0]+'_'+str(years[i]-3)+'.csv')
            path_llb_five_year=os.path.join(path_llb,'sz_llb_'+ls[0]+'_'+str(years[i]-4)+'.csv')
        
        try:
            
            file_fzb=csv.reader(open(path_fzb_year, 'r'))
            file_llb=csv.reader(open(path_llb_year, 'r'))
            file_lrb=csv.reader(open(path_lrb_year, 'r'))
            
        except FileNotFoundError:
            continue

        
    #---------------------------------------取出想分析的数据列进行分析----------------------------------------------------           
        for j_fzb,row_fzb in enumerate(file_fzb):
            fzb_len = j_fzb
        for x in range(0,len(row_fzb)):
            if row_fzb[x] == "":
                row_fzb[x] = '1'
        for j_lrb,row_lrb in enumerate(file_lrb):
            pass
        for x in range(0,len(row_lrb)):
            if row_lrb[x] == "":
                row_lrb[x] = '1'
        for j_llb,row_llb in enumerate(file_llb):
            pass
        for x in range(0,len(row_llb)):
            if row_llb[x] == "":
                row_llb[x] = '1'
        #资产表
        #现金与约当现金 --True  100   1
        gp.loc[[zc[0]],years[i]] = (float(row_fzb[7])+float(row_fzb[21]))/(float(row_fzb[39])+1)*100
        #应收款项    ---True  100   1
        gp.loc[[zc[1]],years[i]] = (float(row_fzb[9])+float(row_fzb[10]))/(float(row_fzb[39])+1)*100
        #存货       ---True 100    1
        gp.loc[[zc[2]],years[i]] = float(row_fzb[16])/(float(row_fzb[39])+1)*100
        #流动资产   ---True   100     1
        gp.loc[[zc[3]],years[i]] = float(row_fzb[20])/(float(row_fzb[39])+1) * 100
        #非流动资产   ---True  100     1
        gp.loc[[zc[4]],years[i]] = float(row_fzb[38])/(float(row_fzb[39])+1) * 100
        #应付款项     ----True   100    1
        gp.loc[[zc[5]],years[i]] = (float(row_fzb[42])+float(row_fzb[43]))/(1+float(row_fzb[39])) * 100
        #流动负债    ----True   100     1
        gp.loc[[zc[6]],years[i]] = float(row_fzb[53])/(1+float(row_fzb[39])) * 100
        #非流动负债   ---True   100     1
        gp.loc[[zc[7]],years[i]] = float(row_fzb[61])/(1+float(row_fzb[39])) * 100
        #股东权益   ---True    100       1
        gp.loc[[zc[8]],years[i]] = float(row_fzb[74])/(1+float(row_fzb[39])) * 100

        #五大财务比率
        #负债占资产比率  --- true   100      1
        gp.loc[[zc[9]],years[i]] = float(row_fzb[62])/(1+float(row_fzb[75])) * 100
        #长期资金占不动产及设备比率 ---true     1
        gp.loc[[zc[10]],years[i]] = (float(row_fzb[74]) + float(row_fzb[61]))/ \
                                    (1+float(row_fzb[26]) + float(row_fzb[27]) + float(row_fzb[28])) * 100
        #流动比率:  ----true     100       1
        gp.loc[[zc[11]],years[i]] = float(row_fzb[20])/(1+float(row_fzb[53])) * 100
        #速动比率    --true    100          1
        gp.loc[[zc[12]],years[i]] = (float(row_fzb[20]) - float(row_fzb[16]) - float(row_fzb[11]))/(1+float(row_fzb[53])) * 100
        #固定资产周转率  --true             1
        gp.loc[[zc[17]],years[i]] = float(row_lrb[8])/(1+float(row_fzb[26]))
        #税前纯益占实收资本比率   - true      1
        gp.loc[[zc[24]],years[i]] = float(row_lrb[29])/(1+float(row_fzb[63]))*100
        #毛利率   --true
        gp.loc[[zc[25]],years[i]] = (float(row_lrb[8]) - float(row_lrb[11]))/(1+float(row_lrb[8]))*100
        #营业利润率  --true
        gp.loc[[zc[26]],years[i]] = (float(row_lrb[23]))/(1+float(row_lrb[8]))*100
        #净利率     -true
        gp.loc[[zc[27]],years[i]] = float(row_lrb[32])/(1+float(row_lrb[8]))*100
        #营业费用率   --true
        gp.loc[[zc[28]],years[i]] = (float(row_lrb[13])+float(row_lrb[14])+float(row_lrb[16]))/(1+float(row_lrb[8]))*100
        #经营安全边际率    --true
        gp.loc[[zc[29]],years[i]] = ((float(row_lrb[23]))/(1+float(row_lrb[8]))*100)/((float(row_lrb[8]) - float(row_lrb[11]))/(1+float(row_lrb[8]))+1)
        #EPS     --true
        gp.loc[[zc[30]],years[i]] = float(row_lrb[32])/(1+float(row_fzb[63]))
        #现金流量比率   ---true  ##16
        gp.loc[[zc[34]],years[i]] = float(row_llb[17])/(1+float(row_fzb[53]))*100
        #现金再投比率    --true   
        gp.loc[[zc[36]],years[i]] = (float(row_llb[17]) - float(row_llb[37]))/(float(row_fzb[39]) - float(row_fzb[53]))*100
        try:
            file_fzb_pre = csv.reader(open(path_fzb_pre_year, 'r'))   
            file_lrb_pre=csv.reader(open(path_lrb_pre_year, 'r'))
        except FileNotFoundError:
            continue
        
        for j_fzb_pre,row_fzb_pre in enumerate(file_fzb_pre):
            pass
        for x in range(0,len(row_fzb_pre)):
            if row_fzb_pre[x] =="":
                row_fzb_pre[x] = '1'
        for j_lrb_pre,row_lrb_pre in enumerate(file_lrb_pre):
            pass
        for x in range(0,len(row_lrb_pre)):
            if row_lrb_pre[x] == "":
                row_lrb_pre[x] = '1'
        #应收款项周转率   --true          
        gp.loc[[zc[13]],years[i]] = float(row_lrb[9])/(1+float(row_fzb[9])+float(row_fzb[10])+float(row_fzb_pre[10])+float(row_fzb_pre[9]))*2
        #存货周转率   --true
        gp.loc[[zc[15]],years[i]] = float(row_lrb[11])/(1+float(row_fzb[16])+ float(row_fzb_pre[16]))*2
        #总资产周转率 --true
        gp.loc[[zc[21]],years[i]] = float(row_lrb[9])*2/(1+float(row_fzb[39])+float(row_fzb_pre[39]))
        #ROA   --true
        gp.loc[[zc[22]],years[i]] = float(row_lrb[32])*2/(1+float(row_fzb[39]) + float(row_fzb_pre[39]))*100
        #ROE   --true
        gp.loc[[zc[23]],years[i]] = float(row_lrb[32])*2/(1+float(row_fzb[74]) + float(row_fzb_pre[74]))*100
        #营收增长率(%)   --true
        gp.loc[[zc[31]],years[i]] = float(row_lrb[8])/(1+float(row_lrb_pre[8]))*100 - 100
        #净利增长率(%)    ---true
        gp.loc[[zc[32]],years[i]] = float(row_lrb[32])/(1+float(row_lrb_pre[32]))*100 - 100
        #净资本增长率(%)    --true
        gp.loc[[zc[33]],years[i]] = float(row_fzb[74])/(1+float(row_fzb_pre[74]))*100 - 100 
            #应收款项周转天数   --true  360
        gp.loc[[zc[14]], years[i]] = (float(row_fzb[9])+float(row_fzb[10])+float(row_fzb_pre[10])+float(row_fzb_pre[9]))*fzb_len*45/(1+float(row_lrb[8]))
                #存货周转天数    --true   360
        gp.loc[[zc[16]],years[i]] = (float(row_fzb[16])+ float(row_fzb_pre[16]))*90*fzb_len/(1+2*float(row_lrb[11]))
                #完整生意周期(天)      --true   360
        gp.loc[[zc[18]],years[i]] = (float(row_fzb[9])+float(row_fzb[10])+float(row_fzb_pre[10])+float(row_fzb_pre[9]))*fzb_len*45/(1+float(row_lrb[8]))+\
                                        (float(row_fzb[16])+ float(row_fzb_pre[16]))*fzb_len*45/(1+float(row_lrb[11]))
        #应付款项周转天数(天)    360
        gp.loc[[zc[19]],years[i]] = (float(row_fzb[42])+float(row_fzb[43])+float(row_fzb_pre[42])+float(row_fzb_pre[43]))*fzb_len*45/(1+float(row_lrb[11]))
        #缺钱天数(天)      360
        gp.loc[[zc[20]],years[i]] = (float(row_fzb[9])+float(row_fzb[10])+float(row_fzb_pre[10])+float(row_fzb_pre[9]))*fzb_len*45/(1+float(row_lrb[8]))+\
                                    (float(row_fzb[16])+ float(row_fzb_pre[16]))*fzb_len*45/(1+float(row_lrb[11])) - \
                                    (float(row_fzb[42])+float(row_fzb[43])+float(row_fzb_pre[42])+float(row_fzb_pre[43]))*fzb_len*45/(1+float(row_lrb[11]))
        try:
            file_fzb_five = csv.reader(open(path_fzb_five_year, 'r'))   
            file_llb_two=csv.reader(open(path_llb_two_year, 'r'))
            file_llb_three=csv.reader(open(path_llb_three_year, 'r'))
            file_llb_four=csv.reader(open(path_llb_four_year, 'r'))
            file_llb_five=csv.reader(open(path_llb_five_year, 'r'))
        except:
            continue
            
        for j_fzb_five,row_fzb_five in enumerate(file_fzb_five):
            pass
        for x in range(0,len(row_fzb_five)):
            if row_fzb_five[x] =="":
                row_fzb_five[x] = '1'
        for j_llb_two,row_llb_two in enumerate(file_llb_two):
            pass
        for x in range(0,len(row_llb_two)):
            if row_llb_two[x] == "":
                row_llb_two[x] = '1'
        for j_llb_three,row_llb_three in enumerate(file_llb_three):
            pass
        for x in range(0,len(row_llb_three)):
            if row_llb_three[x] == "":
                row_llb_three[x] = '1'
        for j_llb_four,row_llb_four in enumerate(file_llb_four):
            pass
        for x in range(0,len(row_llb_four)):
            if row_llb_four[x] == "":
                row_llb_four[x] = '1'
        for j_llb_five,row_llb_five in enumerate(file_llb_five):
            pass
        for x in range(0,len(row_llb_five)):
            if row_llb_five[x] == "":
                row_llb_five[x] = '1'
        s1 = float(row_llb_five[17]) + float(row_llb_four[17]) + float(row_llb_three[17]) + \
            float(row_llb_two[17]) + float(row_llb[17])
        s2 = float(row_llb_five[24]) + float(row_llb_four[24]) + float(row_llb_three[24]) + \
            float(row_llb_two[24]) + float(row_llb[24])
        s3 = float(row_llb_five[20]) + float(row_llb_four[20]) + float(row_llb_three[20]) + \
            float(row_llb_two[20]) + float(row_llb[20])
        s4 = float(row_llb_five[37]) + float(row_llb_four[37]) + float(row_llb_three[37]) + \
            float(row_llb_two[37]) + float(row_llb[37])
        c2 = float(row_fzb_five[16])
        c1 = float(row_fzb[16])
        gp.loc[[zc[35]], years[i]] = s1/(s2-s3+c1-c2+s4)*100
    gp.to_csv('..\\OutFile\\finally\\'+dir+'.csv')
print('完成！')
```
